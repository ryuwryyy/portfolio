```python
import openpyxl
from openpyxl import Workbook
from openpyxl.chart import (RadarChart,Reference)

wb = Workbook()
ws = wb.active

# Excelファイルを開きます
outputFilePath = 'poc.xlsx' # 出力したいExcelファイル名を指定してください
wb = openpyxl.load_workbook(outputFilePath)
sheetNames = wb.sheetnames

# シート取得
ws = wb[sheetNames[0]]

# 行列指定 
labels = Reference(ws, min_col=1, min_row=58, max_row=64)
data = Reference(ws, min_col=2, min_row=74, max_row=80)
# data2 = Reference(ws, min_col=2, min_row=66, max_row=72)

# シートを作成 TODO: 128 -> 要素数
for i in range(128):
    ws1 = wb.create_sheet()
    # シート番号から作業するシートを取得
    ws2 = wb.get_sheet_by_name('Sheet' + str(i+1))

    # TODO 出力するグラフのサイズ
    chtWidth = 60
    chtHeight = 30

    # チャート作成
    chart = RadarChart()
    chart.type = "filled"

    # 一列づつずらす
    data = Reference(ws, min_col=2+(i), min_row=74, max_row=80)
    
    chart.add_data(data, titles_from_data=False)
    chart.set_categories(labels)
    chart.style = 26
    
    # タイトルに数字をつける
    chart.title = ("user" + str(i + 1) + "'s innovation type")
    
    # 場所指定
    ws2.add_chart(chart, "A1")
    
    # 1位と2位の取得

ideal_type_rank1 = []
ideal_type_rank2 = []
ideal_type_rank3 = []
ideal_type_rank4 = []
ideal_type_rank5 = []
ideal_type_rank6 = []
ideal_type_rank7 = []

danger_type_rank1 = []
danger_type_rank2 = []
danger_type_rank3 = []
danger_type_rank4 = []
danger_type_rank5 = []
danger_type_rank6 = []
danger_type_rank7 = []

for j in range(128):    
    # 指定範囲の取得 
    ws3 = wb.get_sheet_by_name('Sheet' + str(j+1))
    
    data2 = []
    for col in ws.iter_cols(min_col=2+j, max_col=2+j, min_row=74, max_row=80, values_only=True):
        data2.append(col)
        
    print("data2: " + str(data2))
#     TODO: 例外処理
#     if data2 == [(0, 0, 0, 0, 0, 0, 0)]:
#         ws3["A17"] = '1位のタイプは' + str(l[0])
#         ws3["A18"] = '2位のタイプは' + str(l[0])
#         ws3["A19"] = 'あなたのタイプは' + str(l[0])
        
    # index を取得 & ソート
    sort_data = tuple(sorted(enumerate(list(data2[0])), key=lambda x: x[1])[5:7])
    print("sort_data:" + str(sort_data))

    rank1 = sort_data[1][0]
    rank2 = sort_data[0][0]
    print("rank1:" + str(rank1))
    print("rank2:" + str(rank2))
    
    type_label = [['機会獲得型', 0],
    ['外交官型', 1],
    ['専門家型', 2],
    ['達成者型', 3],
    ['再定義型', 4],
    ['変容者型', 5],
    ['アルケミスト型', 6]]
    
    # rank
    for l in type_label:
        if l[1] == rank1:
            ws3["A17"] = 'あなたの理想のタイプは' + str(l[0])
            if l[1] == 0:
                ideal_type_rank1.append(rank1)
            elif l[1] == 1:
                ideal_type_rank2.append(rank1)
            elif l[1] == 2:
                ideal_type_rank3.append(rank1)
            elif l[1] == 3:
                ideal_type_rank4.append(rank1)
            elif l[1] == 4:
                ideal_type_rank5.append(rank1)
            elif l[1] == 5:
                ideal_type_rank6.append(rank1)
            elif l[1] == 6:
                ideal_type_rank7.append(rank1)
                
        if l[1] == rank2:
            ws3["A18"] = 'あなたの陥りがちなタイプは' + str(l[0])
            if l[1] == 0:
                danger_type_rank1.append(rank2)
            elif l[1] == 1:
                danger_type_rank2.append(rank2)
            elif l[1] == 2:
                danger_type_rank3.append(rank2)
            elif l[1] == 3:
                danger_type_rank4.append(rank2)
            elif l[1] == 4:
                danger_type_rank5.append(rank2)
            elif l[1] == 5:
                danger_type_rank6.append(rank2)
            elif l[1] == 6:
                danger_type_rank7.append(rank2)
        
        # 最後の傾向
        if rank1 > rank2 and l[1] == rank1:
            ws3["A19"] = 'あなたのタイプは' + str(l[0])
        elif rank1 < rank2 and l[1] == rank2:
            ws3["A19"] = 'あなたのタイプは' + str(l[0])

print("理想のタイプ合計")            
print(len(ideal_type_rank1))
print(len(ideal_type_rank2))
print(len(ideal_type_rank3))
print(len(ideal_type_rank4))
print(len(ideal_type_rank5))
print(len(ideal_type_rank6))
print(len(ideal_type_rank7))

print("陥りがちなタイプ合計")
print(len(danger_type_rank1))
print(len(danger_type_rank2))
print(len(danger_type_rank3))
print(len(danger_type_rank4))
print(len(danger_type_rank5))
print(len(danger_type_rank6))
print(len(danger_type_rank7))

# シート追加


# セルを縦に


wb.save("業務行動分析PoCサーベイ.xlsx")
```
