# random central

```python
%matplotlib inline
import math
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
from pathlib import Path 
from PIL import Image

directory = Path('src/data/preprocess/processed/')

IMAGE_SIZE = 28
height = IMAGE_SIZE
width = IMAGE_SIZE

with tf.Session() as sess:
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(coord=coord)
    
    # escape to name duplicataion 
    count = 0
    for img_list in directory.iterdir():

         if img_list.suffix == '.jpg':
            fpath = str(img_list)
            jpg = tf.read_file(fpath)
            img = tf.image.decode_jpeg(jpg, channels=3)
            
            # central_crop
            central_crop = tf.image.central_crop(img, 0.5)
            img = sess.run(central_crop)
                
            print(img.shape)
            count +=1
            Image.fromarray(np.uint8(img)).save('src/data/preprocess/data_aug/' + img_list.stem +'-random_central'+ str(count) + img_list.suffix)

            plt.imshow(img)
            plt.show()
```

# random_brightness

```python
%matplotlib inline
import math
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
from pathlib import Path 
from PIL import Image

directory = Path('src/data/preprocess/processed/')

IMAGE_SIZE = 28
height = IMAGE_SIZE
width = IMAGE_SIZE

with tf.Session() as sess:
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(coord=coord)
    
    # escape to name duplicataion 
    count = 0
    # how many repeat
    for i in range(3):
        for img_list in directory.iterdir():

             if img_list.suffix == '.jpg':
                fpath = str(img_list)
                jpg = tf.read_file(fpath)
                img = tf.image.decode_jpeg(jpg, channels=3)

                # random_brightness
                brightness = tf.image.random_brightness(img, max_delta=0.75)
                img = sess.run(brightness)

                print(img.shape)
                count +=1
                Image.fromarray(np.uint8(img)).save('src/data/preprocess/data_aug/' + img_list.stem + '-random_bright' + str(count) + img_list.suffix)

                plt.imshow(img)
                plt.show()
```
